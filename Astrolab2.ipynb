```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
Types=np.array([6,11,16,21,26,31,36,41,46,51,56,61,66,71])
Mag=np.array([-5.8,-4.1,-1.1,0.7,2,2.6,3.4,4.4,5.1,5.9,7.3,9.0,11.8,16])
plt.scatter(Types,Mag)
plt.grid(True)
fit=np.polyfit(Types,Mag,3)
fitx=np.arange(0,71,0.1)
def fitf(x):
    return fit[0]*x**3+fit[1]*x**2+fit[2]*x+fit[3]
plt.plot(fitx,fitf(fitx))#Used to check fit to data
```

```python
fitf(9)#Using the function to find the absolute magnitude of a given stellar type
```
