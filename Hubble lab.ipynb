```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
D=np.array([211,542,77.6,506,295])
V=np.array([15.026,42.044,5.992,39.020,21.025])
Verr=np.array([0.161,0.229,.128,.161,.161])
x1=np.arange(0,600,10)
def fit(x):
    return 0.07575*x
```

```python
plt.scatter(D,V, c =['r'])
plt.errorbar(D,V,yerr=Verr,fmt='o',ecolor='black')
plt.plot(x1,fit(x1),'grey')
plt.grid(True)
plt.xlabel("D/MPc")
plt.ylabel("V/Km/s/10^3")
plt.legend(['H=75.76km/s/Mpc'])
```
