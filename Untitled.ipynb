```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
dt=np.array([74.5,50.25,36.65,16.5,7.25])
em=np.array([2.42,0.024,0.01,0.004,0.002])
ec=np.array([21,15.3,10,5.6,2.6])
print(np.polyfit(dt,ec,1))
print(np.polyfit(dt,em,1))
x=np.arange(0,80,1)
```

```python
plt.scatter(dt,em)
plt.errorbar(dt,em,xerr=0.5,yerr=1,fmt='o')
plt.scatter(dt,ec)
plt.errorbar(dt,ec,xerr=0.5,yerr=0.5,fmt='o')
plt.plot(x,0.275*x+0.73)
plt.plot(x,0.031*x-0.673)
plt.grid(True)
plt.xlabel("Temperature Difference (K)")
plt.ylabel("Effiency(%)")
plt.legend(['Efficiency of Carnot','Efficiency Measured'])
plt.title("Graph of Temperature Difference vs Efficiency")
```
