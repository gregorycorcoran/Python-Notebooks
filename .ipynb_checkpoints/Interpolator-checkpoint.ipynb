```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
from astropy.table import Table
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
data=ascii.read(r"C:\Users\gjc10\Downloads\tns_2023jlx_2023-05-29_08-32-51.464012_P60_SEDM_ZTF.ascii")
```

```python
print(np.min(data['col1']),np.max(data['col1']))
```

```python
revdata=Table()
revdata['wav']=np.linspace(np.min(data['col1']),np.max(data['col1']),len(data['col1']))
revdata['f']=np.interp(revdata['wav'],data['col1'],data['col2'])
```

```python
revdata
```

```python
revdata.write(r"C:\Users\gjc10\Downloads\tns_2023jlx_2023-05-29_08-32-51.464012_P60_SEDM_ZTF.txt",\
              format='ascii.fixed_width_no_header',overwrite=True,delimiter=' ')
```

```python
plt.plot(revdata['wav']/1.066,revdata['f'])
plt.grid(True)
plt.axvline(x=6355)
plt.axvline(x=5876)
```
