```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
N=np.array([29,22,11])
H=np.array([0,2.184*10**(-4),4.224*10**(-4)])
plt.scatter(H,np.log(N))
plt.errorbar(H,np.log(N),xerr=0.03*H,yerr=0.05*np.log(N),fmt='o')
print(np.polyfit(H,np.log(N),1,cov=True))
x=np.linspace(0,0.0005,100)
plt.plot(x,-2283*x+3.44)
plt.xlabel('h(m)')
plt.ylabel('Ln(N)')
plt.grid(True)
plt.title('Plot of Ln(N) against h ')
```
