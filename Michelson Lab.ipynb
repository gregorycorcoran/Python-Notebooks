```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
Number = np.array([0,3,6,9,12,15])
Pressure= np.array([0,16,26,42,56,70])
def LinFunc(a,x):
    return a*x
popt,pcov = curve_fit(LinFunc,Number,Pressure)
plt.scatter(Number,Pressure)
plt.errorbar(Number,Pressure,xerr=0,yerr=0.05*Pressure,fmt='o')
plt.plot(Number,LinFunc(Number,popt),color='black')
plt.grid(True)
print("slope error",np.sqrt(pcov[0]))
print("slope",popt)
plt.xlabel('Number of fringes')
plt.ylabel('Pressure (kPa)')
```
