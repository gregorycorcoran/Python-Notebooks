```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import fft, ifft, dct, fftfreq
from scipy.signal import periodogram, blackman
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
path=r'C:\Users\gjc10\Desktop\Fourier'
```

```python
RedData1000=np.loadtxt(path+'\Red_100_1000')
RedData5000=np.loadtxt(path+'\Red_100_5000')
RedData7000=np.loadtxt(path+'\Red_100_7000')
```

```python
plt.plot(RedData5000)
plt.xlim(1000,1500)
plt.xlabel('Voltage samples')
plt.ylabel('Voltage (V)')
plt.grid(True)
```

```python
def Gaussian(x,mu,sig,amp):
    return amp*np.exp((-(x-mu)**2)/(2*sig**2))
```

```python
plt.plot(np.abs(fft(RedData5000,n=8192))/np.max(np.abs(fft(RedData5000,n=8192))))
plt.xlabel('Raw Spectrogram data points')
plt.ylabel('Normalised Intensity')
```

```python
space8192=np.linspace(0,8191,8192)
plt.scatter(2*8192*(6.15)/space8192,np.abs(fft(RedData5000,n=8192))/np.max(np.abs(fft(RedData5000,n=8192))),\
           label='Data points')
plt.xlim(0,1000)
plt.plot(np.linspace(500,800,300),Gaussian(np.linspace(500,800,300),670,45,1),color='r',linestyle='--',\
         label='Manual Gaussian ($\mu=670$nm,$\sigma=45$nm)')
plt.grid(True)
plt.xlabel('Wavelength(nm)')
plt.ylabel('Normalised Intensity')
plt.legend()
```

```python

f5000,pxx5000=periodogram(RedData5000,fs=100/615,nfft=8192,window='blackman')
plt.scatter(2/f5000,pxx5000/np.max(pxx5000),label='Data Points')
rf,drf=curve_fit(Gaussian,2/f5000[1::],pxx5000[1::]/np.max(pxx5000[1::]),p0=[670,30,1])
print(rf)
plt.plot(np.linspace(500,800,300),Gaussian(np.linspace(500,800,300),rf[0],rf[1],2*rf[2]),color='r',\
        linestyle='--',label='Fit($\mu=681$nm,$\sigma=31$nm)')
plt.xlim(0,1000)
plt.grid(True)
plt.xlabel('Wavelength(nm)')
plt.ylabel('Normalised Intensity')
plt.legend()
```
