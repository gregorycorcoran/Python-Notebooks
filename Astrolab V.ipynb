```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.coordinates import SkyCoord
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
Asteroid = SkyCoord(ra=['15:30:38.7','15:30:38.71','15:30:38.72','15:30:38.7','15:30:38.7','15:30:38.69','15:30:38.68'],dec=['+11:14:06.2','+11:14:14.5','+11:14:18.3','+11:15:26.8','+11:15:34.6','+11:15:41.2','+11:15:54.7'],unit=(u.hourangle,u.deg))
Refs = SkyCoord(ra=['15:30:46.56','15:30:39.39','15:30:36.23'],dec=['+11:15:00.8','+11:15:14.2','+11:16:20.8'],unit=(u.hourangle,u.deg))
A= (8.33*10**(-4))#*u.deg#
```

```python
print(Asteroid.ra.deg[1])
```

```python
plt.scatter(Asteroid.ra.deg,Asteroid.dec.deg,marker='x',color='black')
plt.scatter(Refs.ra,Refs.dec,marker='*',s=500,color='r')
plt.errorbar(Asteroid.ra.deg,Asteroid.dec.deg,xerr=A,yerr=A,fmt='o',color='black')
plt.xlim(232.696,232.649)
plt.grid(True,which='both')
plt.text(Asteroid.ra.deg[0]-0.001,Asteroid.dec.deg[0],'B05')
plt.text(Asteroid.ra.deg[1]-0.001,Asteroid.dec.deg[1],'B07')
plt.text(Asteroid.ra.deg[2]-0.001,Asteroid.dec.deg[2],'B08')
plt.text(Asteroid.ra.deg[3]-0.001,Asteroid.dec.deg[3],'B09')
plt.text(Asteroid.ra.deg[4]-0.001,Asteroid.dec.deg[4],'B10')
plt.text(Asteroid.ra.deg[5]-0.001,Asteroid.dec.deg[5],'B12')
plt.text(Asteroid.ra.deg[6]-0.001,Asteroid.dec.deg[6],'B14')
plt.text(Refs.ra.deg[0]-0.001,Refs.dec.deg[0],'R1')
plt.text(Refs.ra.deg[1]-0.001,Refs.dec.deg[1],'R2')
plt.text(Refs.ra.deg[2]-0.001,Refs.dec.deg[2],'R3')
plt.xlabel('RA (°)')
plt.ylabel('Dec (°)')
```

```python
Para=SkyCoord(ra=['15:30:37.72','15:30:38.69'],dec=['11:15:36.1','11:15:41.2'],unit=(u.hourangle,u.deg))
```

```python
plt.scatter(Para.ra.deg,Para.dec.deg,marker='x',color='black')
plt.errorbar(Para.ra.deg,Para.dec.deg,xerr=A,yerr=A,fmt='o',color='black')
plt.scatter(Refs.ra.deg,Refs.dec.deg,marker='*',color='r',s=500)
plt.xlim(232.696,232.649)
plt.text(Refs.ra.deg[0]-0.001,Refs.dec.deg[0],'R1')
plt.text(Refs.ra.deg[1]-0.001,Refs.dec.deg[1],'R2')
plt.text(Refs.ra.deg[2]-0.001,Refs.dec.deg[2],'R3')
plt.text(Para.ra.deg[0]-0.001,Para.dec.deg[0],'ATEAST')
plt.text(Para.ra.deg[1]+0.006,Para.dec.deg[1],'ATWEST')
plt.xlabel('RA (°)')
plt.ylabel('Dec (°)')
plt.grid(True,which='Both')
```
