```python
page=rq.get("https://www.hines.com/properties?country_name_in=Ireland&isSidebarOpen=true&loc=-59.17592824927136%7C-169.45312500000003%7C79.49665183507658%7C179.29687500000003")
soup = bs(page.content, "html.parser")
print(soup)
```

```python
results=soup.find("div",class_="l-property-results-list js-property-results-list")
#print(results)
res2=results.find_all("a",class_="property-result")
#print(len(res2))
#print(res2[0]['href'])
sublinks=[res['href'] for res in res2]
#print(sublinks)
fulllinks=["https://www.hines.com"+x for x in sublinks]
print(fulllinks[714])
```

```python
prop_page=rq.get(fulllinks[25])
prop_soup=bs(prop_page.content,"html.parser")
```

```python
res3=prop_soup.find("li",class_="small-list__item")
print(res3.contents)
address=""
for x in res3.contents:
    address+=x
    if x != res3.contents[-1]:
        address+=", "
    else:
        #print("Test")
        break
print(address)
```

```python
#Full tool
#names=[]
#addresses=[]

for link in fulllinks[714:]:
    address=""
    name=link.split("/")[-1]
    names.append(name)
    page=rq.get(link)
    soup=bs(page.content,"html.parser")
    results=soup.find("li",class_="small-list__item")
    for x in results.contents:
        address+=x
        if x != results.contents[-1]:
            address+=", "
        else:
            #print("Test")
            break
    addresses.append(address)
    sleep(10)
#print(names)
print(addresses)
```

```python
#names.pop(-1)
print(names[715])
print(addresses[715])
names.pop(715)
addresses.pop(715)
print(len(names))
print(len(addresses))
```

```python
addresses2=[x.strip("\n") for x in addresses]
```

```python
dict= {'name':names,'address':addresses2}
df=pd.DataFrame(dict)
df
```
