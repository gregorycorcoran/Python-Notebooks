```python
def BiasAvg(dataset):
    Sum=np.zeros([2048,2098],dtype=float)
    for bias in dataset:
        data=fits.getdata(bias)
        Sum+=data.astype(float)
    return Sum/len(dataset)
    
```

```python
def MasterFlats(flatset,MBias):
    Sum=np.zeros([2048,2098],dtype=float)
    for flat in flatset:
        flatdata=fits.getdata(flat)
        Sum+=flatdata-MBias
    return Sum/np.mean(Sum[974:1074,999:1099])
MFlat=MasterFlats(flats,MasterBias)
```

```python
def Reductor(Imset,mbias,mflat):
    RedSet=[]
    for Im in Imset:
        imdata=fits.getdata(Im)
        #Crop gets rid of dead pixels
        final=(imdata[::,50:2097].astype(float)-mbias[::,50:2097].astype(float))/(mflat[::,50:2097].astype(float))
        RedSet.append(final)
    return RedSet
```

## Attempt 1 of making an all in one function for this ##

```python
def MagFinder(imageset,DataSet):
    #Preamble for use in later arrays
    exptime=[]
    fluxes=[]
    bkgfluxes=[]
    times=[]
    ApAreas=[]
    AnAreas=[]
    m_std=[15.8,15.1,13.7,13.7]
    ZP=[]
    mag_NN=[]
    dmag_NN=[]
    ZPAVG=[]
    for data in DataSet:
        #To get exp time and time
        exp=float(fits.getheader(data)['EXPTIME'])
        exptime.append(exp)
        time=fits.getheader(data)['UT']
        times.append(time)
    for image in imageset:
        A=[0,0,0]
        B=[0,0,0]
        C=[0,0,0]
        D=[0,0,0]
        NN=[0,0,0]
        
        bkgstd=mad_std(image)
        daofinds= DAOStarFinder(fwhm=6,threshold=5*bkgstd)
        sources=daofinds.find_stars(image)
        for source in sources:
            #The idea of this being that given we know roughly where each star is we can find the sources
            #This doesnt work well if the star tracker isnt perfect, which became an issue
            #Essentially if no star was found it would just use the [0,0,0]
            
            if source['xcentroid']>1337.5 and source['xcentroid']<1339 and source['ycentroid']>1150 and\
            source['ycentroid']<1153:
                A[0]=source['id']
                A[1]=source['xcentroid']
                A[2]=source['ycentroid']
                #print(A)
            if source['xcentroid']>1293.5 and source['xcentroid']<1295 and source['ycentroid']>1496.5 and\
            source['ycentroid']<1498:
                B[0]=source['id']
                B[1]=source['xcentroid']
                B[2]=source['ycentroid']
                #print(B)
            if source['xcentroid']>1066.5 and source['xcentroid']<1068.5 and source['ycentroid']>1481.5 and\
            source['ycentroid']<1483.5:
                C[0]=source['id']
                C[1]=source['xcentroid']
                C[2]=source['ycentroid']
                #print(C)
            if source['xcentroid']>1378.5 and source['xcentroid']<1380.5 and source['ycentroid']>838.5 and\
            source['ycentroid']<840.5:
                D[0]=source['id']
                D[1]=source['xcentroid']
                D[2]=source['ycentroid']
                #print(D)
            if source['xcentroid']>1445 and source['xcentroid']<1447 and source['ycentroid']>1145 and\
            source['ycentroid']<1147:
                NN[0]=source['id']
                NN[1]=source['xcentroid']
                NN[2]=source['ycentroid']
                #print(NN)
        #Initially these were in the if statements which meant I was not actually seeing that
        #they were not found
        if A[0]==0:
                print('A Not Found')
        if B[0]==0:
                print('B Not Found')
        if C[0]==0:
                print('C Not Found')
        if D[0]==0:
                print('D Not Found')
        if NN[0]==0:
                print('NN Not Found')
        
        positions=np.transpose([[A[1],B[1],C[1],D[1],NN[1]],[A[2],B[2],C[2],D[2],NN[2]]])
        apertures=CircularAperture(positions,r=6)
        annulus_apertures=CircularAnnulus(positions,r_in=10,r_out=15)
        AllAps=[apertures,annulus_apertures]
        photo_table=aperture_photometry(image,AllAps)
        #print(photo_table)
        bkgmean=photo_table['aperture_sum_1']/annulus_apertures.area
        bkgtotal=bkgmean*apertures.area
        finalsum=photo_table['aperture_sum_0']-bkgtotal
        fluxes.append(finalsum)
        bkgfluxes.append(bkgtotal)
        ApAreas.append(apertures.area)
        AnAreas.append(annulus_apertures.area)
        print(finalsum[4])
        
    #print(fluxes[0][0:3])
    #print(fluxes[1])
    #Fluxes shape is (18,5)
    #print(type(fluxes[0][0]))
    #print(type(exptime[0]))
    #Doing fluxes[0][0:3] gets the flux of the first 4 objects in the first image
    for i in range(0,len(imageset)):
        #To get zeropoints and mag
        ZPa=m_std[0]+2.5*np.log10(fluxes[i][0]/exptime[i])
        ZPb=m_std[1]+2.5*np.log10(fluxes[i][1]/exptime[i])
        ZPc=m_std[2]+2.5*np.log10(fluxes[i][2]/exptime[i])
        ZPd=m_std[3]+2.5*np.log10(fluxes[i][3]/exptime[i])
        ZPavg=np.average([ZPa,ZPb,ZPc,ZPd])
        print(ZPa,ZPb,ZPc,ZPd,ZPavg)
        m_NN=ZPavg-2.5*np.log10(fluxes[i][4]/exptime[i])
        #print(ApAreas)
        #print(AnAreas)
        #print(bkgfluxes[i][4])
        SN=(fluxes[i][4])/np.sqrt(fluxes[i][4]+(ApAreas[0]*(1+(ApAreas[0]/AnAreas[0])))*\
                                (bkgfluxes[i][4]+3.37**2))
        sig=2.5*np.log10(1+(1/SN))
        mag_NN.append(m_NN)
        dmag_NN.append(sig)
        ZPAVG.append(ZPavg)
        
        
    return mag_NN,times,dmag_NN,ZPAVG
```

## Attempt 2 ##

```python
def MagFinderV2(ImageSet,DataSet):
    exptime=[]
    fluxes=[]
    bkgfluxes=[]
    times=[]
    ApAreas=[]
    AnAreas=[]
    m_std=[15.8,15.1,13.7,13.7]
    ZP=[]
    mag_NN=[]
    dmag_NN=[]
    ZPAVG=[]
    Positions=[]
    LimMags=[]
    for data in DataSet:
        exp=float(fits.getheader(data)['EXPTIME'])
        exptime.append(exp)
        time=fits.getheader(data)['UT']
        times.append(time)
    bkgstd=mad_std(ImageSet[0])
    daofinds= DAOStarFinder(fwhm=6,threshold=5*bkgstd)
    sources=daofinds.find_stars(ImageSet[0])
    #This is the solution to the previous problem that was used
    #The stars are found in the first image which finds them all, the aperatures placed then
    #Then simply the flux in each aperature is taken in each image
    for row in [27,40,41,48,49]:
        #This was taken from previously
        Positions.append([sources['xcentroid'][row-1],sources['ycentroid'][row-1]])
    #This means
    #0 is D
    #1 is NN
    #2 is A
    #3 is C
    #4 is B
    apertures=CircularAperture(Positions,r=6)
    annulus_apertures=CircularAnnulus(Positions,r_in=10,r_out=15)
    AllAps=[apertures,annulus_apertures]
    for image in ImageSet:
        photo_table=aperture_photometry(image,AllAps)
        #print(photo_table)
        bkgmean=photo_table['aperture_sum_1']/annulus_apertures.area
        bkgtotal=bkgmean*apertures.area
        finalsum=photo_table['aperture_sum_0']-bkgtotal
        fluxes.append(finalsum)
        bkgfluxes.append(bkgtotal)
        ApAreas.append(apertures.area)
        AnAreas.append(annulus_apertures.area)
    for i in range(0,len(ImageSet)):
        ZPd=m_std[3]+2.5*np.log10(fluxes[i][0]/exptime[i])
        ZPa=m_std[0]+2.5*np.log10(fluxes[i][2]/exptime[i])
        ZPc=m_std[2]+2.5*np.log10(fluxes[i][3]/exptime[i])
        ZPb=m_std[1]+2.5*np.log10(fluxes[i][4]/exptime[i])
        ZPavg=np.average([ZPa,ZPb,ZPc,ZPd])
        #print(ZPa,ZPb,ZPc,ZPd,ZPavg)
        m_NN=ZPavg-2.5*np.log10(fluxes[i][1]/exptime[i])
        #print(ApAreas)
        #print(AnAreas)
        #print(bkgfluxes[i][4])
        SN=(fluxes[i][1])/np.sqrt(fluxes[i][1]+(ApAreas[0]*(1+(ApAreas[0]/AnAreas[0])))*\
                                (bkgfluxes[i][1]+3.37**2))
        sig=2.5*np.log10(1+(1/SN))
        mag_NN.append(m_NN)
        dmag_NN.append(sig)
        ZPAVG.append(ZPavg)
    #Need to now make limiting magnitude function
    #Can go through each image finding all the sources
    #Using the ZPavg from each image find all the magnitudes
    #Then find minimum from each image
    for number,image in enumerate(ImageSet):
        #print(number)
        bkgstdlim=mad_std(image)
        daofinds= DAOStarFinder(fwhm=6,threshold=5*bkgstdlim)
        sourceslim=daofinds.find_stars(image)
        poslim=np.transpose([sourceslim['xcentroid'],sourceslim['ycentroid']])
        apertures=CircularAperture(poslim,r=6)
        annulus_apertures=CircularAnnulus(poslim,r_in=10,r_out=15)
        AllApslim=[apertures,annulus_apertures]
        phot_tablim=aperture_photometry(image,AllApslim)
        #print(phot_tab0)
        bkgmeanlim=phot_tablim['aperture_sum_1']/annulus_apertures.area
        bkgtotlim=bkgmeanlim*apertures.area
#print(bkgtot0)
        finalsumlim=phot_tablim['aperture_sum_0']-bkgtotlim
#print(finalsum0)
        mags=ZPAVG[number]-2.5*np.log10(finalsumlim/exptime[number])
    #Will get negative logs which cause there to be a point missing
    
        LimMags.append(np.max(mags))
        #print(np.max(mags))
    return mag_NN,times,dmag_NN,ZPAVG,LimMags
    
         
```

```python
finalmags,times,dfinalmags,ZPavg,LimMags=MagFinderV2(RedSet,DataSet)
```
