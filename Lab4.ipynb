```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
datax1 = np.array([0,5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80,86])
```

```python
datay1 = np.array([1,0,1,1,0.8,0.37,1,1,0.4,0.18,0.6,0.7,0.51,0.3,0.04,0.11,0.06,0.09])
```

```python
plt.plot(datax1,datay1)
plt.xlabel("angle(°)")
plt.ylabel("Meter Reading")
plt.grid(True)
plt.errorbar(datax1,datay1,xerr=2,yerr=0.02,ecolor="b")
# d= 9cm
```
