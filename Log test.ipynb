```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
times=[]
modifiers=[]
modifier=0.01
time=0
while modifier<1:
    modifier*=1.09854
    time+=0.04
    if modifier>1:
        modifier=1
    modifiers.append(modifier)
    times.append(time)
plt.scatter(times,modifiers)
plt.yscale('log')
print(times[-1])
print(len(times))
```
